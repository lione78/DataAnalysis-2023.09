### 2.10 서울시 구별 CCTV 현황 분석
    - 서울시 자치구 연도별 CCTV 설치 현황
    - 서울시 주민등록인구(구별) 통계

```python
import numpy as np
import pandas as pd
```

1. CCTV 데이터

```python
cctv = pd.read_csv('data/서울시 자치구 년도별 CCTV 설치 현황_230630기준.csv', index_col='구분', encoding='EUC-KR', thousands=',')
cctv.head(3)
```

```python
cctv.drop(columns=['Unnamed: 0', '2023년', '총계'], inplace=True)
cctv.drop(['계'], inplace=True)            # '계' 행 지우기
cctv.head(3)
```

```python
# 결측치가 있는지 확인.
cctv['2014년 이전'].isna().sum()
```

```python
# 결측치를 0으로 대체하고, 타입을 정수로 변경
cctv['2014년 이전'] = cctv['2014년 이전'].fillna(0).astype(int)
cctv.head(3)
```

```python
cctv['CCTV댓수'] = cctv.sum(axis=1)
cctv.head(3)
```

```python
# cctv['2020년' : '2022년']은 행 인덱스로 인식
cctv['최근증가율'] = (cctv.loc[:,'2020년':'2022년'].sum(axis=1) / cctv.CCTV댓수 * 100).round(2)
cctv.head(3)
```

```python
cctv.drop(columns=['2014년 이전'] + '2014년	2015년	2016년	2017년	2018년	2019년	2020년	2021년	2022년'.split(),inplace=True)
cctv
```

```python
# '중 구'의 blank 없애기
cctv.index = cctv.index.str.replace(' ','')
cctv.index[1]
```

```python
cctv.index.name = '구별'
cctv.head(3)
```

#### 2. 인구 데이터 정리하기

```python
pop = pd.read_csv('data/서울시 구별 주민등록인구.csv', skiprows=1)
pop.head()
```

```python
pop = pop[['동별(2)', '계 (명)', '한국인 (명)', '등록외국인 (명)', '65세이상고령자 (명)']]
pop.head(3)
```

```python
pop.columns = ['구별', '인구수', '내국인', '외국인', '고령자']
```

```python
# 첫번째 행 지우기
pop.drop([0], inplace=True)
pop.head(3)
```

```python
pop.set_index('구별', inplace=True)
pop.head()
```

```python
# 외국인 비율, 고령자 비율
pop['외국인 비율'] = (pop.외국인 / pop.인구수 * 100).round(2)
pop['고령자 비율'] = (pop.고령자 / pop.인구수 * 100).round(2)
pop.head(3)
```

### 3. CCTV 데이터와 POP 데이터 합치기

```python
cctv_pop = pd.merge(cctv, pop, left_index=True, right_index=True)
cctv_pop.head(3)
```

```python
cctv_pop.shape
```

```python
cctv_pop.index.name = '구별'
```

```python
cctv_pop.to_csv('data/서울시 구별 CCTV 인구 현황.csv')
```
