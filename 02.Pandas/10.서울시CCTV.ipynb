### 2.10 서울시 구별 CCTV 현황 분석
    - 서울시 자치구 연도별 방범용 CCTV 운영 현황
    - 서울시 주민등록인구

```python
import numpy as np
import pandas as pd
```

```python
cctv = pd.read_csv('data/서울시 자치구 연도별 방범용 CCTV 운영 현황_230630기준.csv', encoding='EUC-KR', index_col='구분', thousands=',')
cctv
```

```python
cctv.drop(columns=['Unnamed: 0', '2023년'], inplace=True)
cctv.head(3)
```

```python
cctv['CCTV댓수'] = cctv.sum(axis=1)
cctv.head(3)
```

```python
cctv['최근증가율'] = (cctv.iloc[:,5:-2].sum(axis=1) / cctv.CCTV댓수 * 100).round(2)
cctv.head(3)
```

```python
cctv.drop(columns={'2017년','2018년','2019년','2020년','2021년','2022년'},inplace=True)
cctv
```

```python
cctv.index.name = '구별'
cctv.head(3)
```

#### 2. 인구 데이터 정리하기

```python
pop = pd.read_csv('data/서울시 구별 주민등록인구.csv', skiprows=1)
pop.head()
```

```python
pop = pop[['동별(2)', '계 (명)', '한국인 (명)', '등록외국인 (명)', '65세이상고령자 (명)']]
pop.head(3)
```

```python
pop.columns = ['구별', '인구수', '내국인', '외국인', '고령자']
```

```python
# 첫번째 행 지우기
pop.drop([0], inplace=True)
pop.head(3)
```

```python
pop.set_index('구별', inplace=True)
pop.head()
```

```python
# 외국인 비율, 고령자 비율
pop['외국인 비율'] = (pop.외국인 / pop.인구수 * 100).round(2)
pop['고령자 비율'] = (pop.고령자 / pop.인구수 * 100).round(2)
pop.head(3)
```

### 3. CCTV 데이터와 POP 데이터 합치기

```python
cctv_pop = pd.merge(cctv, pop, left_index=True, right_index=True)
cctv_pop.head(3)
```

```python
cctv_pop.shape
```

```python
cctv_pop.to_csv('data/서울시 구별 CCTV 인구 현황.csv')
```
