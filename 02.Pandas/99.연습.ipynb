```python
cctv = pd.read_csv('data/서울시 자치구 연도별 방범용 CCTV 운영 현황_230630기준.csv', encoding='EUC-KR', index_col='No.')
cctv.head()
```

```python
cctv_1 = pd.DataFrame()
for col in cctv.columns:
    if col == '구분':
        cctv_1[col] = cctv[col]
    elif col != '구분':
        cctv_1[col] = cctv[col].apply(lambda x: int(str(x).replace(',','')))
cctv_1.reset_index(inplace=True)
cctv_1
```

```python
cctv_1.drop(columns=['level_0', 'index'], inplace=True)
```

```python
cctv_1.loc[:, '2015년':'2023년'] = cctv_1.loc[:, '2015년':'2023년'].astype(int)
cctv_1.set_index(keys=['No.', '구분'], inplace=True)
cctv_1
```

```python
cctv_1.loc['년도별 합계', :]
```

```python
cctv_1['구별 합계'] = cctv_1.sum(axis=1).astype(int)
cctv_1.loc['년도별 합계'][:] = cctv_1.sum().astype(int)
cctv_1
```
