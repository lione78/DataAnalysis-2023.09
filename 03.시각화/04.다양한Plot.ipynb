### 2. Bar Plot

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
```

```python
plt.figure(figsize=(5, 3))      # figsize =(x, y), y값을 줄이는 것이 중요
plt.bar([0, 1, 2], [2, 3, 1])
plt.title('Bar Chart')
plt.xticks([0, 1, 2], ['가', '나', '다']), plt.yticks([1, 2, 3])        # [0, 1, 2] -> ['가', '나', '다'] 대응
plt.xlabel('가나다'), plt.ylabel('빈도수')
plt.show()
```

```python
s = pd.Series(np.random.rand(10), index=list('abcdefghij'))
plt.bar(s.index, s.values)
plt.title('Bar Chart')      # plt.xticks([]), plt.grid()
plt.show()
```

```python
s.plot(kind='bar'), plt.show()
```

```python
colors = ['blue', 'orange', 'green', 'red', 'purple', 'brown', 'pink', 'gray', 'olive', 'cyan']
plt.bar(s.index, s.values, color=colors)
plt.title('Bar Chart')      # plt.xticks([]), plt.grid()
plt.show()
```

```python
s.index = ['종로구', '중구', '용산구', '성동구', '광진구', '동대문구', '중랑구', '성북구', '강북구', '도봉구']
plt.barh(s.index, s.values, color=colors)
plt.title('Horizontal Bar Chart')
plt.show()
```

### 3. 파이 차트

```python
# 0도 부터 시계 반대 방향으로
ratio = [10, 15, 20, 28]        # 합이 100이 아니어도 됨.
labels = ['감', '귤', '밤', '배']
plt.pie(ratio, labels=labels, autopct='%.1f%%')     # 0.1까지
plt.title('파이 차트')
plt.show()
```

```python
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
explode = (0, 0.1, 0, 0)        # 0.1 정도 튀어 나옴.
plt.pie(ratio, labels=labels, autopct='%.1f%%', colors=colors, explode=explode, shadow=True, startangle=90)
plt.title('파이 차트')
plt.show()
```

### 5. 산점도(Scatter plot)

```python
import seaborn as sns
iris = sns.load_dataset('iris')
X, Y = iris.sepal_length, iris.sepal_width
plt.scatter(X, Y)
plt.xlabel('sepal_length'), plt.ylabel('sepal_width')
plt.title('산점도 - iris sepal length vs width')
plt.show()
```

```python
colors = ['red'] * 50 + ['green'] * 50 + ['blue'] * 50
plt.scatter(X, Y, color=colors)
plt.xlabel('sepal_length'), plt.ylabel('sepal_width')
plt.title('산점도 - iris sepal length vs width')
plt.show()
```

```python
size = 500 * np.random.rand(150)
plt.scatter(X, Y, c=colors, alpha=0.5, s=size)
plt.title('Bubble chart')
plt.show()
```

### 7. Box Plot

```python
# setosa 품종의 각 피처(ssl, ssw, spl, spw)
ssl = iris[iris.species == 'setosa'].sepal_length
ssw = iris[iris.species == 'setosa'].sepal_width
spl = iris[iris.species == 'setosa'].petal_length
spw = iris[iris.species == 'setosa'].petal_width
```

```python
plt.boxplot((ssl, ssw, spl, spw),
            labels=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'])
plt.title('Setosa 품종의 Feature')
plt.show()
```
