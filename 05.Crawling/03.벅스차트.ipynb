```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url = 'https://music.bugs.co.kr/chart'
res = requests.get(url)
print(res.text)
```

```python
soup = BeautifulSoup(res.text, 'html.parser')
```

```python
trs = soup.select('.list.trackList.byChart > tbody > tr')
trs[:2]
```

```python
tr = trs[0]
tr
```

```python
rank = int(tr.select_one('.ranking > strong').get_text().strip())
rank
```

```python
image = tr.select_one('.thumbnail > img')['src']
image
```

```python
title = tr.select_one('.title').get_text().strip()
title
```

```python
artist = tr.select_one('.artist').get_text().strip()
artist
```

```python
album = tr.select_one('.album').get_text().strip()
album
```

```python
data = []
for tr in trs:
    rank = int(tr.select_one('.ranking > strong').get_text().strip())
    image = tr.select_one('.thumbnail > img')['src']
    title = tr.select_one('.title').get_text().strip()
    artist = tr.select_one('.artist > a').get_text().strip()
    album = tr.select_one('.album').get_text().strip()
    data.append({'순위': rank, 'image': image, '곡이름' : title, '아티스트': artist, '앨범': album})
data
```

```python
df = pd.DataFrame(data)
df.head()
```
