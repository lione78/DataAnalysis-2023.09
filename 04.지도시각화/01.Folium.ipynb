- map 객체 만들기

```python
map = folium.Map(location=[37.53372, 126.905131], zoom_start=14)       # 영등포역 위도, 경도
map
```

- tileset 선택

```python
# default tileset: Open Street Map
# Stamen Toner : 단계 구분도(Choropleth)를 그릴 때 사용
map = folium.Map(location=[37.53372, 126.905131], zoom_start=14, tiles='Stamen Toner')
map
```

- Marker 추가

```python
map = folium.Map(location=[37.52272, 126.905131], zoom_start=14)
folium.Marker(
    location=[37.51528, 126.9075611],            # WGS84 좌표계: 위도(latitude), 경도(longitude)
    tooltip='영등포역',
    popup=folium.Popup('서울시 영등포구 경인로 846', max_width=200)
).add_to(map)
folium.Marker(
    location=[37.52272, 126.905131],
    tooltip='영등포시장역',
    popup=folium.Popup('서울시 영등포구 양산로 200', max_width=200)
).add_to(map)
map
```

- 다양한 Marker + 제목 달기

```python
map = folium.Map(location=[37.52272, 126.905131], zoom_start=14)
folium.Marker(
    location=[37.51528, 126.9075611],            # WGS84 좌표계: 위도(latitude), 경도(longitude)
    tooltip='영등포역',
    popup=folium.Popup('서울시 영등포구 경인로 846', max_width=200)
).add_to(map)
folium.Marker(
    location=[37.52272, 126.905131],
    tooltip='영등포시장역',
    popup=folium.Popup('서울시 영등포구 양산로 200', max_width=200),
    icon=folium.Icon(color='green')
).add_to(map)
folium.Circle(
    location=[37.532003, 126.914072],
    tooltip='국회의사당',
    radius=100, color='crimson', fill=True,
    popup=folium.Popup('서울특별시 영등포구 여의도동 1', max_width=200)
).add_to(map)
folium.CircleMarker(
    location=[37.525317, 126.896606],
    tooltip='영등포구청',
    radius=15, color='#3186cc', fill=True,
    popup=folium.Popup('서울특별시 영등포구 당산동3가 385-1', max_width=200)
).add_to(map)
title_html = '<h3 align="center" style="font-size:20px">학원근처 유명 지점</h3>'
map.get_root().html.add_child(folium.Element(title_html))
map
```

```python
import pandas as pd
```

```python
df = pd.read_csv('data/영등포소재 지하철역2.csv')
df
```

```python
map = folium.Map(location=[37.52272, 126.905131], zoom_start=14)
for i in df.index:
    folium.Marker(
        location = [df['위도'][i],df['경도'][i]],
        tooltip = df['이름'][i],
        popup = folium.Popup(df['주소'][i], max_width=200)
    ).add_to(map)
map
```
