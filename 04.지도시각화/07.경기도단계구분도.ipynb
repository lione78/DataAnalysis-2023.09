```python
import pandas as pd
```

```python
geo_data1 = json.load(open('data/seoul_geo_simple.json', encoding='UTF-8'))
```

```python
df1 = pd.read_csv('../02.Pandas/data/서울시 구별 CCTV 인구 현황.csv', index_col='구별')
```

```python
map1 = folium.Map(location=[37.55, 126.98], zoom_start=11, tiles='Stamen Toner')
folium.Choropleth(
    geo_data=geo_data1,
    data=df1.CCTV댓수,
    columns=[df1.index, df1.CCTV댓수],
    fill_color='Greens',
    key_on='feature.id'
).add_to(map1)
html_title = '<h3 align="center" style="font-size: 15px">구별 CCTV 현황</h3>'
map1.get_root().html.add_child(folium.Element(html_title))
map1
```

```python
import numpy as np
def get_text_location(geo_str):
    gu_dict = {}
    for gu in geo_str['features']:
        for coord in gu['geometry']['coordinates']:
            geo = np.array(coord)
            gu_dict[gu['id']] = [np.mean(geo[:,1]), np.mean(geo[:,0])]
    return gu_dict
```

```python
gu_dict = get_text_location(geo_data1)
gu_dict
```

```python
for gu_name, gu_values in gu_dict.items():
    folium.map.Marker(
        location=[gu_values[0], gu_values[1]],
        icon=folium.DivIcon(icon_size=(80, 20),icon_anchor=(20,0),
                            html=f'<span style="font-size: 15px">{gu_name}</span>')
    ).add_to(map1)
map1
```
